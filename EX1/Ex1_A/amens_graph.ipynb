```python
from main import *
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import numpy as np
import math

df = separate_cols()["amenities"]
```

```python
l= {}
for ams in df:
    ams=eval(ams)
    for am in ams:
        try: l[am]+=1
        except: l[am]=1

l=dict(sorted(l.items(), key=lambda item: item[1], reverse=False))

# lim=400
# l = {k: l[k] for k in l.keys() if l[k]>lim}
```

```python
total_count_uses_by_amens = [0 for i in range(len(l.keys()))]
for i,k in enumerate(l.keys()):
    total_count_uses_by_amens[i]+=\
        total_count_uses_by_amens[i-1]+l[k]

plt.title("how much each amen type adds to use")
plt.xlabel("amens type by #")
plt.ylabel("contrib of xth to total amens count")
plt.ylim(0.0,0.1)
plt.xlim(len(l.keys())-200,len(l.keys()))
plt.plot([i/total_count_uses_by_amens[-1] for i in total_count_uses_by_amens])
plt.show()
```

```python
for_len=3000#max(l.values())
amens_with_above_i_use = [0 for i in range(for_len)]
for k in l.keys():
    for i in range(for_len):
        if l[k]<=i: amens_with_above_i_use[i]+=1

plt.title("how many amens were used less than x")
plt.ylabel("% of amens with less than x")
plt.ylabel("count of amens used in total")
# plt.xlim(0,500)
plt.ylim(0.95,1)
plt.plot([i/amens_with_above_i_use[-1] for i in amens_with_above_i_use])
plt.show()
```
